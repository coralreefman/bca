# Approaches in the Arts

![Image](images/bense-beuys-1970.png)  
In the famous 1970 debate in Düsseldorf, Joseph Beuys and Max Bense clashed over fundamentally different visions of what art should be and do.  

Joseph Beuys is of course the famous German post-WWII artist, theorist that became know for his large scale installations and social 'interventions'. A well-know statement of Beuys was that ‘Jeder Mensch ist ein Künstler’ and ‘Jeder Künstler ist ein Mensch,’ which underlines his humanistic view on art production.  

Max Bense was a German philosopher, aesthetic theorist, and pioneer of information aesthetics. His work bridges *semiotics, cybernetics, and modernist art*, seeking a rational, scientific grounding for aesthetic judgment.  

## Joseph Beuys’ Perspective
Beuys argued that art is a transformative social force—a “social sculpture” in which every individual has creative potential to affect societal change. For him, art transcends formal aesthetics and is inherently political, intuitive, and performative. He embraced an expansive view where art is intertwined with everyday life and the process of change, emphasizing spontaneity, personal expression, and the power of ritualistic acts. His approach was less concerned with rigid, objective criteria and more with the experiential and transformative qualities of art.  

## Max Bense’s Perspective
In contrast, Max Bense championed a rational, semiotic, and formally grounded approach to aesthetics. He maintained that art should be evaluated by clear, objective criteria and that its meaning emerges from structured, communicative processes. Bense’s view was that a rigorous, critical analysis—drawing on principles of logic, information theory, and semiotics—could demystify art, making its value accessible and intelligible. He was skeptical of mystical or overly subjective interpretations, arguing instead for an art that is both measurable and conceptually clear.  

The debate was essentially a clash between Beuys’ view of art as a fluid, socially engaged, and transformative practice, and Bense’s insistence on an objective, rational framework for understanding and judging art. This confrontation not only reflected their personal philosophies but also symbolized broader tensions in contemporary art and aesthetics between *intuition and intellect*, *subjectivity and objectivity*, and art’s role as *a tool for social change versus a domain of disciplined inquiry*.

## Max Bense's philosophy: Information Aesthetics and concrete art manifesto.  

An early promoter of this say *rational art* approach is the German artist Max Bense. In the 1950s he formulated his Information Aesthetics.  

### 1. Foundations: Aesthetics as a Science

Concept & Meaning:  
*Aesthetic Formalism*  
Beauty resides in the /formal/ properties of an artwork (structure, symmetry, proportion), not in emotional or representational content.  

*Information Theory*  
Inspired by Claude Shannon, Bense treats an artwork as a *message* that carries /information/. The amount and organization of this information determine its aesthetic value.  

*Objective Measurement*  
He proposes quantitative criteria (e.g., entropy, complexity) to evaluate artworks, aiming for a *normative, inter-subjectively verifiable*  aesthetic judgment.             |


*Key Quote* – “Art is a /communication of information/ whose value lies in the /efficiency/ of its coding.”

### 2. The “Aesthetic Measure”

Bense introduced a *formulaic approach*  to assess visual works, especially abstract and geometric art:

[ \\text{Aesthetic Value} = f(\\text{Complexity}, \\text{Order}, \\text{Redundancy}) ]

- *Complexity* – Number of distinct elements or variations.
- *Order* – Degree of regularity, symmetry, or predictable pattern.
- *Redundancy* – Repetition that reduces informational entropy, enhancing recognizability.

He argued that *optimal art*  balances high complexity (richness) with sufficient order (coherence) and controlled redundancy (memorability).

### 3. Semiotics & the “Sign-System”

Bense treated an artwork as a *sign-system*  composed of:

1. *Syntax*  – Formal relations among visual elements (lines, shapes, colors).
2. *Semantics*  – Potential meanings that can be inferred, though he downplayed subjective interpretation.
3. *Pragmatics*  – Contextual effects (viewer’s cultural background, medium).

His semiotic model stresses that *meaning arises from structural relations* , not from the artist’s intent alone.

### 4. Relation to Modernist Movements

Movement & Bense’s Position  
*Concrete Art*  (Theo van Doesburg, Max Bill)  
Ideal exemplar: pure geometry, no illusion, maximal informational clarity.  

*Constructivism*  
Appreciated its engineering-like rigor; saw it as a step toward “objective art”.  

*Abstract Expressionism*  
More skeptical—high emotional content conflicted with his formalist, information-centric criteria.  

Bense collaborated with concrete artists (e.g., *Max Bill*) and helped formulate the manifesto of the *International Society of Concrete Art* (1960s). (see below)

### 5. The “Aesthetic Object” vs. “Aesthetic Experience”

- *Object-Centric*: The artwork itself possesses measurable properties; the viewer’s reaction is secondary.
- *Experience-Centric*: While acknowledging the viewer, Bense maintains that a /trained observer/ can discern the objective qualities through systematic analysis.

Thus, *education* becomes crucial: learning to “read” the formal language of art enables reliable aesthetic judgments.

### 6. Influence of Cybernetics & Systems Theory

- *Feedback Loops*: Bense likened the artistic process to a *closed system*  where the creator adjusts forms based on informational criteria, akin to a cybernetic controller.
- *Self-Organization*: He praised artworks that exhibit emergent order without external direction, seeing this as a manifestation of natural informational laws.

### 7. Critiques & Legacy

Criticism & Bense’s Response / Legacy  
*Reductionism* – Reducing beauty to numbers ignores affect.  
He acknowledged affect but argued it can be /derived/ from structural harmony.  

*Subjectivity Neglected* – Viewers differ wildly.  
He proposed a /trained/  aesthetic competence that minimizes variance.  

*Technological Bias* – Over-reliance on digital metrics.  
Later scholars adapted his ideas to *digital art*, algorithmic composition, and *generative design*.

Today, Bense’s ideas surface in:

- *Computational aesthetics*  (algorithmic evaluation of visual complexity).
- *Design theory*  (principles of balance, proportion, and information density).
- *Digital humanities*  (quantitative analysis of visual corpora).

### 8. Quick Take-aways for Practitioners

1. *Measure Form* – Use ratios, symmetry counts, and entropy calculations to assess visual balance.
2. *Seek Optimal Complexity* – Too simple = boring; too chaotic = incomprehensible. Aim for a middle ground.
3. *Educate the Eye* – Train yourself to read structural cues; this sharpens aesthetic judgment.
4. *Leverage Technology* – Software can compute information metrics (e.g., image entropy) aligning with Bense’s methodology.

*Bottom line:*  
Max Bense attempted to turn aesthetics into a disciplined, quantitative science, arguing that the beauty of an artwork can be objectively measured through its informational structure, balance of complexity and order, and the semiotic relations of its visual signs. His legacy lives on in contemporary computational aesthetics and design theory.

## Influence on the concrete art movement
As we have just seen, Max Bense was a core contributor to the concrete art manifesto.

Here are a few of the core tenets of the manifesto:

**Art Must Be “Concrete”**

*“Concrete art is not abstract art; it is an art that does not refer to anything outside itself.”*

- *Self-referentiality* : The artwork’s meaning is contained entirely within its own formal elements (line, plane, colour, rhythm).
- *No Symbolic Reference* : No allegory, narrative, or depiction of external objects is permitted.

**Formal Purity & Rational Construction**

- *Geometric Basis* : Only elementary geometric forms (points, lines, planes, circles, polygons) are admissible.
- *Mathematical Order* : Proportions, symmetry, and modular systems must be derived from explicit mathematical relationships (e.g., golden ratio, Fibonacci, lattice grids).
- *Deterministic Process* : The creation process follows a reproducible set of rules; randomness is allowed only when it can be expressed as a rule (e.g., algorithmic iteration).

**Information-Theoretic Aesthetic**

- *Information Content*: An artwork is a /message/ ; its aesthetic value correlates with the amount of organized information it conveys.
- *Optimal Complexity*: Beauty arises from a balance between *complexity* (variety of elements) and *order* (regularity, predictability). Excessive redundancy lowers information value; excessive chaos raises entropy beyond comprehension.
- *Quantifiable Criteria* : Though the manifesto does not provide explicit formulas, it anticipates later attempts (by Bense himself) to measure aesthetic value via entropy, redundancy, and structural complexity.

**Universality & Internationalism**

- *Language-Free* : Because the visual code is purely geometric, it transcends linguistic and cultural barriers.
- *Global Network* : The Society calls for an international exchange of works, exhibitions, and publications, fostering a worldwide “visual lingua franca”.

**Ethical Dimension**

- *Art as Scientific Inquiry*: Concrete art is positioned as a /research activity/ that contributes to the understanding of perception, cognition, and the laws governing visual information.
- *Responsibility to Truth* : Artists must avoid deception; the work must be honest about its construction and not hide “hidden meanings”.

### Reception & Impact

Aspect, Immediate Reaction and Long‑Term Influence:

*Critical Reception*  
Mixed: praised for rigor, criticized as “cold” and “mechanical”.  
Became a reference point for later *minimalist*, *systems*, and *computational*  art movements.  

*Exhibitions*  
First major show: /Concrete Art – International Exhibition/  (Zurich, 1961).  
Annual /Concrete Art Biennale/  (Stuttgart, 1963‑1975) cemented a network of artists across Europe and Japan.  

*Academic Legacy*  
Bense’s later works ( /Aesthetik und Informationslehre/ ) built directly on the manifesto’s ideas.  
Influences contemporary *generative art*, *algorithmic composition*, and *digital aesthetics*  (e.g., software that evaluates image entropy).  

*Political/Ideological*  
Positioned as apolitical, yet its emphasis on universal rationality resonated with certain *modernist* and *technocratic*  currents. | Occasionally invoked in debates about **“objective” versus “subjective” art criticism.  

### Critical Perspectives

1. *Reductionism* – Critics argue that reducing art to geometry and information neglects affect, cultural context, and the embodied experience of viewing.
2. *Accessibility* – The strict formal constraints can be perceived as elitist, limiting participation to those with strong technical training.
3. *Historical Myopia* – By rejecting all representational content, the manifesto sidelines non-Western visual traditions that employ abstraction differently.
4. *Technological Determinism* – The reliance on “information theory” presumes a linear progression from scientific discovery to artistic practice, a premise contested by later post-structuralist thinkers.

Nevertheless, supporters contend that the manifesto *provides a clear methodological framework* that enables reproducibility, cross-cultural dialogue, and a rigorous discourse on visual perception.

### The Manifesto in Practice – How Artists Apply It

Step and Typical Workflow (based on the manifesto):  

*1. Define a Formal System*  
Choose a set of geometric primitives (e.g., squares, circles) and a mathematical relationship (e.g., golden ratio).  

*2. Compute Information Content*  
Use software (MATLAB, Python, Processing) to calculate entropy of the pattern; adjust parameters to reach a target complexity.  

*3. Produce the Work*  
Render the design via hand-drawing, printmaking, or digital fabrication (laser-cut, CNC).  

*4. Document Construction*  
Record all dimensions, ratios, and algorithmic steps in a catalogue entry (as required by the ISCA archival standards).  

*5. Exhibit & Exchange*  
Submit the piece to a concrete-art exhibition or online platform that adheres to ISCA’s peer-review guidelines.  

### Summary
1. *Concrete art = self-contained visual language* ; no external references.
2. *Form follows explicit mathematical rules*; the artist is a /designer of information/ .
3. *Aesthetic value ≈ optimal balance of complexity & order* (early precursor to today’s computational aesthetics).
4. *Internationalism* : The manifesto seeks a universal visual code that transcends culture.


# Jack Burnham's Systems Esthetics

![image](images/Jack_Burnham_with_one_of_his_light_sculptures.jpg)  

Jack Burnham argues that art in the late-1960s is moving from a focus on isolated, finished objects toward a systems-oriented aesthetic in which the artwork is understood as a network of relationships, processes, and contexts. He frames this shift as a new cultural paradigm comparable to Thomas Kuhn’s scientific revolutions: the old “major paradigm” of autonomous art objects is being replaced by a broader, technology-driven “systems” paradigm.
Below are the core ideas Burnham develops, together with a few of his most striking quotations.

*1. From Objects to Systems*  
	• Old paradigm: Art is a unique, finite object (painting, sculpture) that exists apart from its surroundings.  
	• New paradigm: Art is an environment or system—a set of interacting components (people, materials, information, energy) whose meaning emerges from those relationships.
“The specific function of modern didactic art has been to show that art does not reside in material entities, but in relations between people and between people and the components of their environment.”
“We are now in transition from an object-oriented to a systems-oriented culture. Here change emanates, not from things, but from the way things are done.”

*2. Technological & Socio-Technical Drivers*  
	• The rise of information processing, automation, and large-scale engineering forces artists to confront the same decision-making frameworks used in the Pentagon, industry, and government.  
	• Economists like J.K. Galbraith anticipate a “technocracy” where aesthetic judgment becomes part of policy and resource management.  
“In the emergent ‘superscientific culture’ long-range decision-making and its implementation become more difficult and more necessary… Judgment demands precise socio-technical models.”  
“The major tool for professionally defining these concerns is systems analysis… Systems analysts are not cold-blooded logicians; the best have an ever-expanding grasp of human needs and limitations.”

*3. Redefining the Artist’s Role*  
	• The artist becomes a quasi-political provocateur who must dissolve the traditional separation between art and technology.  
	• Artists such as Duchamp, Warhol, Robert Morris, Carl Andre, Dan Flavin, and Hans Haacke exemplify this shift: they foreground process, information flow, and environmental interaction over static visual form.  
“In an advanced technological culture the most important artist best succeeds by liquidating his position as artist vis-à-vis society.”  
“The systems approach goes beyond a concern with staged environments and happenings; it deals in a revolutionary fashion with the larger problem of boundary concepts.”  

*4. The Aesthetic of Process & Information*  
	• Process (fabrication, kinetic motion, environmental response) supersedes the final object.  
	• Information—whether transmitted via telephone, sensors, or data streams—becomes a legitimate aesthetic element.  
“For systems, information, in whatever form conveyed, becomes a viable aesthetic consideration.”  
“The range of outside factors affecting it, as well… its own radius of action, reach beyond the space it materially occupies. It thus merges with the environment in a relationship that is better understood as a ‘system.’”  

*5. Implications for Culture & Politics*  
	• The systems aesthetic suggests that artistic decisions are also ethical and ecological choices, influencing resource use, environmental stewardship, and societal organization.  
	• By treating art as a decision-making system, the aesthetic becomes a tool for confronting global challenges (energy, waste, social interaction).  
“These decisions—whether they are made concertedly or not—control the quality of future life on the earth.”  
“The systems hierarchy of values… all living situations must be treated in the context of a systems hierarchy of values.”  

*6. Legacy & Outlook*  
	• Burnham predicts that the systems aesthetic will become the dominant framework for art, superseding formalist and post-formalist styles.  
	• He links this shift to earlier avant-garde movements (Constructivism, Productivism) while stressing that the new paradigm is rooted in contemporary technology and socio-technical organization.  

“The emergence of a ‘post-formalist esthetic’ may seem to some to embody a kind of absolute philosophy… yet it is more likely that a ‘systems aesthetic’ will become the dominant approach to a maze of socio-technical conditions rooted only in the present.”

*Key Take-aways*  
1. Art is no longer about isolated objects; it is about systems of relationships, processes, and information.
2. Technological and socio-technical forces (automation, data, large-scale engineering) drive this aesthetic shift.
3. Artists must act as mediators between culture and technology, dissolving the boundary between creator and engineer.
4. Process, environmental interaction, and informational exchange are now central aesthetic criteria.
5. The systems aesthetic carries ethical weight, influencing how societies allocate resources and shape the future.

Burnham’s essay thus reframes the role of art from a decorative, object-focused practice to a critical, systemic practice that participates directly in shaping the technological and ecological fabric of contemporary life.

# the computational turn or a new weltbild

![image](images/network-topologies.png)

Is this essay Yuk Hui makes an argument for a view of the world from a cultural computational perspective. He compares this with the historical mechanical world view and how this has changed late 20th Century. He concludes with a discussion of the network as a technological and cultural entity.

"Today we have to ask, what is the world picture of our time? There has been a shift from a mechanised world picture, based in analogue models of scientific thought, to one formed through the digital network based on patterns and repetitions of digital matter. It is a specific image of networks, which characterises the computational turn."

"I will propose that two apparently distinct domains, namely communication networks and language, converge and fulfill this image of networks."

"An artwork is not an object anymore, but a network. It is beyond the correlation between time and space, since it is discursive; it follows the logic of language, which is in turn only possible through network technology."

This 'Weltbild' (in Heidegger's wording) that Hui argues about here is a continuation of thought put forward by Jack Burnham where artworks are systems of relationships, processes, and information.

Summarized by Lumo:
- *From Tool to World‑Picture*   
  Hui argues that computation is no longer a neutral instrument but has become the /primary way/ we conceive, describe, and act upon the world. The “computational turn” therefore reshapes our /Welt­bild/  (world‑picture) in the same way earlier paradigms (mechanical, linguistic) did.
  
- *Heidegger’s World‑Picture Revisited*   
  Building on Heidegger’s notion of the /Welt­bild/ —the world “conceived and grasped as a picture”—Hui shows that modern technology has turned the world into an image that can be captured, stored, and processed. The computational turn extends Heidegger’s critique of modern technology by making the picture itself algorithmic and networked.
  
- *From Mechanical to Networked Imagery*   
  The historic shift from the *mechanical world‑picture* (Newtonian physics, deterministic machines) to a *networked world‑picture*  is traced through the rise of digital communication, the Internet, and the semantic web. Networks replace the static, ordered mechanisms of the past with dynamic, distributed topologies (centralised, decentralised, distributed).
  
- *Network as a Discursive, Linguistic Structure*   
  Hui links the technical architecture of networks (packet‑switching, scale‑free graphs) with the /inferential model/ of human communication (coding/decoding, relevance theory). The network thus becomes a /discursive/  structure: a logical language instantiated in hardware that shapes cultural dynamics.
  
- *Altermodern and the Disorientation of Time/Space*   
  Drawing on Nicolas Bourriaud’s *Altermodern* , Hui suggests that the computational turn produces a new cultural epoch characterized by:
  
  - *Disorientation* : traditional ties between time and place dissolve; events can be linked across disparate temporalities.
  - *Heterochrony* : multiple, overlapping chronologies coexist, enabled by networked media.
  - *Nomadic Art*: artworks become /networks/  rather than objects, reflecting the fluid, relational logic of the digital milieu.  
  
- *Philosophical Tension: Ontology vs. Epistemology*   
  The essay highlights a persistent tension:
  
  - *Ontological* (Being) versus *Ontic*  (entities) – Heidegger’s distinction blurs when the world itself is rendered as data.
  - *Metaphysical Reflection*  is still required; the computational turn offers powerful possibilities but also risks a “techno‑logo‑centrism” that reduces everything to calculable forms.  
  
- *Conclusion: A Call for New Metaphysics*   

Hui does not dismiss the computational turn; instead, he urges a fresh metaphysical inquiry that can accommodate the networked, discursive world‑picture without surrendering the question of Being.

Quotes:
“In the altermodern, the disorientation of time and space is celebrated; the network reconciles the spatial and the temporal into a single narrative.”  

“We are witnessing tremendous efforts in turning objects into data… the intensive increase in the amount of data creates a digital milieu which we can hardly escape.”  

“The network is a symptom: ‘Network Fever.’” – Mark Wigley (cited)  

“Computation is not anymore a method, methodology, or tool but, more fundamentally, the way we see and act in the world.”  

The essay’s core argument: the computational turn has forged a new, network‑based Welt­bild that reshapes ontology, culture, and artistic practice, demanding fresh philosophical reflection.

![image](images/jacquard-loom.jpg)  

The Jacquard Loom was not only a driver of the Industrial Revolution it also kickstarted the Information Age with the mechanical processing of information through the use of punchcards, a chain of cardboard cards punched with holes that determines which cords of the fabric should be raised with each pass of the shuttle. This proto-digital system used binary code (hole – no hole) to control the behaviour of individual threads in the loom (up or down) and allowed unskilled workers to weave complex patterns. While the binary principle (hole – no hole) still underpins modern day information technology, the threads in the loom also bear striking resemblance to the physical characteristics of the microcircuits that process digital information. Modern microchips can have up to 100 layers.

Summary of Artificial Art by Remko Scha (Jacquard Lecture, 1991)  
(recommended to read the whole article on the drive)

Remko Scha examines the relationship between digital computers, artificial intelligence (AI), and the future of artistic practice. He argues that, while computers excel at formal, symbol-manipulation tasks, they are still ill-suited to the perceptual, associative processes that underlie human cognition—and consequently to the creative decisions of an artist. Because current AI models (especially “expert systems”) are narrowly scoped and lack the rich, flexible perception required for genuine artistic judgment, the prospect of a fully autonomous “artificial artist” is unrealistic for the foreseeable future.  

Nevertheless, Scha highlights the productive role computers can play when we **re-frame** artistic creation:  

1.  **Chance art & generative randomness** – Using mathematically defined spaces of visual possibilities, a program can select random samples, yielding unexpected and unconventional results that differ from human-driven designs.
2.  **Formal visual algebras** – To generate “arbitrary paintings,” one needs a language that encodes elementary shapes and operations, allowing systematic exploration of a combinatorial visual space.
3.  **Computational strengths** – Computers outperform humans in exhaustive search, large-scale calculation, and systematic variation, which can lead to novel, surprising solutions (e.g., circuit layout, chess endgames).

Scha situates this approach historically (references to “chance art,” Duchamp’s readymades, and earlier generative programs such as Harold Cohen’s AARON) and argues that a **division of labor** —humans defining visual vocabularies and constraints, machines sampling from the resulting space— could extend artistic practice beyond the limits of individual imagination.  

He also reflects on broader aesthetic implications:  

-   The spectator, not the creator, ultimately interprets the work, echoing Duchamp’s view that “the spectator makes the picture.”
-   By treating all objects as potential aesthetic material, the traditional hierarchy of “high art” collapses, opening a “Copernican turn” in which art is no longer centered on a singular author but on systematic exploration of visual grammars.

In conclusion, Scha predicts that **well-designed generative algorithms** will increasingly produce surprising, unconventional artworks, while human artists will shift toward defining the underlying visual language and curating the output rather than crafting each piece manually. The paper thus frames computers not as replacements for artists but as **new tools for expanding the combinatorial horizon of visual creation.**
