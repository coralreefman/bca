# Terminal Basics +

The gateway to power over your machine.  
Open a terminal application on your computer.

You will see a prompt like this `>`  
Where you can type commands.  

Type:  
`cd ~` + enter  

Now type:  
`ls` + enter  

It prints a list of the files and directories in your home folder.  

`ls -la` for a listing of metadata; creation date/time, file size, permissions, etc. on the files and directories.  

For example List a directory:  
`ls Documents`  

You always work in a directory in the file/directory structure on your drive. Which so called *working directory* you are currently working in, can be found out with `pwd`:  
`pwd`  
`/home/myusername`  

You can change the working directory with change directory or:  
`cd`  
For example make your Documents folder your working dir:  
`cd Documents`  
And see a listing of its contents.

Now if you want to move up in the hierarchy, use the double dot:  
`cd ..`   

You can also explicitly go to any directory on the drive by using path notation. The root or beginning of the hierarchy starts with a forward slash `/`, followed by a directory name or names separated by `/`s like so:  
`cd /home/username/Documents`  

Note: a shortcut notation is available to go directly to your home folder by means of the `~`, like we did at the start of this tutorial:
`cd ~`  

Or, just list your home directory without going there:  
`ls ~`  

Now, try the following command:

### navigation, super important  
*history*  
The terminal keeps a history of the commands that you have used. This is super useful as it allows to select a previous typed command by just using the UP/DOWN arrow keys.  
Hit UP/DOWN arrows to move through the history of previous commands.  

*command completion*  
Also the terminal is able to complete a partly typed command,so you don't have to type out all commands entirely.
Hit TAB to complete a command or path.  

Try it out, you see, you don't have to type everything out all the time; make use of history and command completion! Computers are here to help, not torture you.

### installing programs
Sometimes you needs to install programs or data at a restricted location on your drive which requires you to gain super user powers.  
The `sudo` command allows you to do this:  
`sudo apt install library`  
=> be very careful though, you can severly fuck things up using the sudo command.  

In general with Python we work in a **clean and safe environment** that can be activated with -as we use here- conda. This is a shielded off area on your drive where you can safely experiment.  

To create a conda virtual environment:  
`conda create --name bca`  

You then need to activate the newly created conda virtual environment:  
`conda activate bca`  

Now you can mess things up safely :)

While in the conda environment you can install software through packages like so:  
`conda install package-name`  

Conda also has build in help:  
`conda -h`

### fun
Inspect the contents of any file on your drive:  
`hexdump some-file-name`  

It shows the actual bytes in the file in hexadecimal notation.  
Now do this and add the `-C` switch or argument:  
`hexdump -C some-file-name`  

See what happens? Yes, everything is just binary data; texts, images, video, etc. Does this mean we can  play a text document as audio? Sure:  
`aplay some-document`
